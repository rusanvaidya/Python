All classes have a function called __init__(), which is always executed when the class is being initiated.

Use the __init__() function to assign values to object properties, or other operations that are necessary to do when the object is being created:

```python
class A:
    def __init__(self, name, age):
        self.fullname = name
        self.age = age
```

```python
class B(A):
    def __init__(self, name, age, address):
        super().__init__(name, age)
        self.address = address
        
    def show(self):
        print(self.fullname ," aged ", self.age, " lives in ", self.address)
```

```python
b = B('Hari', 22, 'Bktp')
```

```python
b.show()
```

```python
class Birds:
    def __iter__(self):
        self.bird = ('peigon', 'sparrow', 'crow', 'eagle', 'hawk', 'vulture')
        self.index = 0
        self.len = len(self.bird)
        return self
    
    def __next__(self):
        if self.index < self.len:
            x = self.bird[self.index]
            self.index += 1
            return x
        else:
            raise StopIteration

mybird = Birds()
bird = iter(mybird)
print(next(bird))
print(next(bird))
print(next(bird))
print(next(bird))
print(next(bird))
print(next(bird))
# print(next(bird))
```
