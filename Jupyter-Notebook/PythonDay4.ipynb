```python
import pandas as pd
```

### Pandas Series

```python
# defining pandas series through list

a = ['a', 'e', 'i', 'o', 'u']
s = pd.Series(a)
print(s)
#  notice that index is numeric value starting from 0
```

```python
# defining the index value for the series
s = pd.Series(a, index=['v1', 'v2', 'v3', 'v4', 'v5'])
print(s)
```

```python
# defining the pandas series through dictionary
animal = {'Land':'Lion', 'Air':'Eagle', 'Water':'Shark'}
s = pd.Series(animal)
print(s)
# notice that the key value is the index and value is value in pandas series
```

### Replaceing the value using mean median or mode

```python
x = df['Speed'].mean()
new_df = df.fillna(x, inplace=True)
```

```python
x = df['Speed'].median()
new_df = df.fillna(x, inplace=True)
```

```python
x = df['Speed'].mode()[0]
new_df = df.fillna(x, inplace=True)
```
