```python
arr = np.array([9, 7, 7, 3, 9, 7, 7, 7, 3, 9])
random.shuffle(arr)
print(arr)
```

```python
print(random.permutation(arr))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.displot(arr, kind='kde')
plt.show()
```

```python
sns.histplot(arr)
plt.show()
```

```python
x_norm = random.normal(loc=1, scale=100, size=(100))

print(x_norm)
sns.displot(x_norm, kind='kde')
```

```python
sns.displot(random.normal(loc=50, scale=7, size=(100)), label='normal', kind='kde')
plt.show()
```

```python
sns.displot(random.binomial(n=100, p=0.5, size=100), label='binomial', kind='kde')
```

```python
sns.displot(random.poisson(lam=100, size=100), label='poisson', kind='kde')
```

```python
sns.displot(random.uniform(size=100), label='uniform', kind='kde')
```

```python
sns.displot(random.logistic(loc=50, scale=7, size=100), label='logistic', kind='kde')
```

```python
print(random.multinomial(n=6, pvals=[1/6, 1/6, 1/6, 1/6, 1/6, 1/6]))
# random rolls e.g dice rolls
```

```python
sns.displot(random.exponential(scale=7, size=100), kind='kde')
```

```python
sns.displot(random.chisquare(df=1, size=100), kind='kde')
```

```python
sns.displot(random.rayleigh(size=100), kind='kde')
```

```python
sns.displot(random.pareto(a=1, size=100), kind='kde')
```

```python
sns.displot(random.zipf(a=2, size=100), kind='kde')
```
