# MATPLOTLIB

```python
import matplotlib.pyplot as plt
import numpy as np

amplitude = []
phase = []

phase_val=0
amp_val = 0
max_amp = 10
amplitude.append(amp_val)
phase.append(phase_val)
while phase_val!=2:
    phase_val = phase_val+1/4
    amp_val = max_amp*np.sin(np.pi*phase_val)
    amplitude.append(amp_val)
    phase.append(phase_val)

print(amplitude)
print(phase)
plt.plot(phase, amplitude)
plt.ylabel('Amplitude')
plt.xlabel('Phase')
plt.grid()
plt.show()
```

```python
plt.plot(phase, amplitude, 'o')
plt.ylabel('Amplitude')
plt.xlabel('Phase')
plt.grid()
plt.show()
```

```python
plt.plot(phase, amplitude, marker='o', ms=10, mec='r', mfc='y')
plt.ylabel('Amplitude')
plt.xlabel('Phase')
plt.grid()
plt.show()
# other markers can be used are '*', '.' , '_', '1', '2', etc.
# ms = width of marker
# mec = edge color of marker
# mfc = color of marker
```

```python
plt.plot(phase, amplitude, 'o:r')
plt.ylabel('Amplitude')
plt.xlabel('Phase')
plt.grid()
plt.show()
# o:r -> o : marker ; r : red
```

```python
plt.plot(phase, amplitude, 'o-')
plt.ylabel('Amplitude')
plt.xlabel('Phase')
plt.grid()
plt.show()
# o- -> 'o': marker ;  - :line
```

```python
plt.plot(phase, amplitude, 'o--')
plt.ylabel('Amplitude')
plt.xlabel('Phase')
plt.grid()
plt.show()
# o-- -> 'o' : marker ; -- : dashed line
```

```python
plt.plot(phase, amplitude, 'o:')
plt.ylabel('Amplitude')
plt.xlabel('Phase')
plt.grid()
plt.show()
# ':'  : dotted line
```

```python
plt.plot(phase, amplitude, 'o-.')
plt.ylabel('Amplitude')
plt.xlabel('Phase')
plt.grid()
plt.show()
# -. : dashed + dotted
```

```python
plt.plot(phase, amplitude, 'o:y')
plt.ylabel('Amplitude')
plt.xlabel('Phase')
plt.grid()
plt.show()
```

```python
plt.plot(phase, amplitude, linestyle='dotted')
plt.ylabel('Amplitude')
plt.xlabel('Phase')
plt.grid()
plt.show()
# linestyle dotted
```

```python
# linestyle can be also wirtten as ls
plt.plot(phase, amplitude, ls='dashed')
plt.ylabel('Amplitude')
plt.xlabel('Phase')
plt.grid()
plt.show()
```

```python
font1 = {'family':'serif','color':'blue','size':15}
font2 = {'family':'serif','color':'darkred','size':15}

# color can also be written as c

plt.subplot(1,2,1)
plt.plot(phase, amplitude, color='#2d7b20', linewidth=10)
plt.title('Sine Wave', fontdict=font1)
plt.ylabel('Amplitude', fontdict = font2, loc='top')
plt.xlabel('Phase', fontdict = font2, loc='right')
plt.grid()
plt.show()

plt.subplot(1,2,2)
plt.plot(phase, amplitude)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

amplitude = []
phase = []
amplitude2 = []
phase2 = []

phase_val=0
amp_val = 0
max_amp = 10
amplitude.append(amp_val)
phase.append(phase_val)
while phase_val!=2:
    phase_val = phase_val+1/4
    amp_val = max_amp*np.sin(np.pi*phase_val)
    amplitude.append(amp_val)
    phase.append(phase_val)


plt.scatter(phase, amplitude)

phase_val=0
amp_val = 0
max_amp = 10
amplitude2.append(amp_val)
phase2.append(phase_val)
while phase_val!=2:
    phase_val = phase_val+1/4
    amp_val = max_amp*np.cos(np.pi*phase_val)
    amplitude2.append(amp_val)
    phase2.append(phase_val)
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige"])

plt.scatter(phase2, amplitude2, c=colors)
plt.ylabel('Amplitude')
plt.xlabel('Phase')
plt.grid()
plt.show()
```

```python
x = np.array(['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I'])
plt.bar(x, np.abs(amplitude2))
plt.show()
```
